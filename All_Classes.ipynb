```python
# write a python fucntion that a string as argument
# function analyze string: len,alphabets,spaces,numbers,upper,lower,
# vowels,words,sentence,paragraph
```

```python
txt = '''hciowdkjwedadqwdjhasdqwed edjkjwedbwejdhasjk hdedwehjdvqwedhbqwjhdvbqwdwwehjdvwedhbqakdd qwhhdbqwkdqw 263 vjhui
sdjhcbsdcsdcsd
djchsdkjcsdjcbsd
jhsdcsd
dasjh82y2'''




def strint_data(txt):
    
    length = len(txt)
    
    vow = 'aeiouAEIOU'
    alphabets = 0
    spaces = 0
    numbers = 0
    lower_case = 0
    upper_case = 0
    others = 0
    for char in txt:
        if char.isalpha():
            alphabets+=1
        elif char.isspace():
            spaces+=1
        elif char.isnumeric():
            numbers+=1
        elif char.islower():
            lower_case +=1
        elif char.isupper():
            upper_case +=1
        elif char in vow:
            vow+=1
        else:
            others+=1
    print(f"""
        Text length : {length}
        
        
        
        lower case : {lower_case}
        uppercase : {upper_case}
        alphabets : {alphabets}
        vowels : {vow}
        others : {others}
    
    
    """)
        
        
        
            
    
```

```python
strint_data(txt)
```

```python
# even num sum
def even_num(*nums):
    result= 0
    for n in nums:
       if n%2==0:
           result+=n
    return result
```

```python
from ABC import ABC , abstractmethod
class Computer(ABC):
    @abstractmethod
    def process(self):
        pass
    @abstractmethod 
    def greet(self):
        print("Hello")

# to make an class abstract
# it must inherit the ABC class from abc module
# it must have atleast a abstract method
# which is defined using a decorator @abstractmethod
```

```python
com1 = Computer()
```
